```python
def create_country_leaderboard(tips):
    return (
        tips.groupby('Gender')[['Total_Bill','price_per_person']]
        .sum()
        .rename(index=str.lower)
        .reset_index()
        .sort_values(by='Tips')
    )
```

```python
def top_5_payer_name(tips):
    tips.dropna(subset=['Payer Name'], inplace=True)
    tips.sort_values(by=['Total_Bill'],ascending=False,inplace=True)
    return tips
```
